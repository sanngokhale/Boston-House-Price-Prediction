```python
bs_X = boston.drop(['MEDV'], axis = 1)
bs_Y = boston['MEDV']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(bs_X, bs_Y, random_state = 101, test_size = 0.51)
```

```python
X_tr = X_train.fillna(0)
X_ts = X_test.fillna(0)
```

```python
bs_X.describe()
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_tr, y_train)
```

```python
print(reg.coef_)
```

```python
y_pred = reg.predict(X_ts)
print(y_pred)
```

```python
print(y_test)
```

```python
#mean squared error
print(np.mean((y_pred-y_test)**2))
```

```python
#check accuracy with mean squared error
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_test, y_pred))
```

```python
boston1 = boston1.fillna(0)
#reg1 = LinearRegression()
bs_X_pred = boston1
#reg1.fit(bs_X_pred, y_train)
y_pred2 = reg.predict(bs_X_pred)
print(y_pred2)
```

```python
#creating submission.csv file
pred = pd.DataFrame(y_pred2)
datasets = pd.concat([boston1.Id, pred], axis = 1)
datasets.columns = ['Id', 'MEDV']
datasets.to_csv('submit.csv', index=False)
```
